VISUALIZATION



```python
# Confirmed Cases by Region

plt.figure(figsize=(20,6))  # Better dimensions
region_confirmed = df.groupby('Region')['Confirmed'].sum().sort_values(ascending=False)
region_confirmed.plot(kind='bar', color='skyblue')  # Optional: blue for confirmed
plt.xlabel('Region')
plt.ylabel('Total Confirmed Cases')
plt.title("Total Confirmed Cases by Region")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Total Deaths by Region

plt.figure(figsize=(20,6))  # Better height for visibility
region_death = df.groupby('Region')['Deaths'].sum().sort_values(ascending=False)
region_death.plot(kind='bar', color='tomato')  # Optional: color for impact
plt.xlabel('Region')
plt.ylabel('Total Deaths')
plt.title("Total Deaths by Region")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Recovered Cases by Region

plt.figure(figsize=(20,6))  # Better size for readability
region_recovered = df.groupby('Region')['Recovered'].sum().sort_values(ascending=False)
region_recovered.plot(kind='bar', color='seagreen')  # Optional: green color for recovery
plt.xlabel('Region')
plt.ylabel('Total Recovered Cases')
plt.title("Total Recovered Cases by Region")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
region_summary = df.groupby("Region")[["Confirmed", "Deaths", "Recovered"]].sum()
region_summary.sort_values(by="Confirmed", ascending=False).head(10)
```

```python
import plotly.express as px

# Assign the top 10 regions summary to the variable top10
top10 = region_summary.sort_values(by="Confirmed", ascending=False).head(10)
top10_reset = top10.reset_index()  # Reset index to have Region as a column

fig = px.bar(top10_reset, x='Region', y=['Confirmed', 'Deaths', 'Recovered'],
             title="Top 10 Regions: Confirmed, Deaths & Recovered Cases",
             labels={'value':'Number of Cases', 'Region':'Region'},
             barmode='group',
             height=500)

fig.show()
```

Conclusion:

* The data analysis shows the distribution of COVID-19 confirmed cases, deaths, and recoveries by region.
* The US, Spain, and Italy are the top affected regions in terms of confirmed cases and deaths.
*  Visualizing these metrics helps understand the impact and recovery trends, providing valuable insights for further monitoring and decision-making.
